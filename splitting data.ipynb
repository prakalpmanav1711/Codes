```python
car_misss = pd.read_csv("car-sales-extended-missing-data.csv")
car_misss
```

```python
car_misss.isna().sum()
```

```python
car_misss.dropna(subset="Price",inplace=True)
car_misss.isna().sum()
```

```python
X_misss = car_misss.drop("Price",axis=1)
Y_miss = car_misss["Price"]
```

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

cat_imputer = SimpleImputer(strategy="constant",fill_value="missing")
door_imputer = SimpleImputer(strategy="constant",fill_value=4)
numerical_imputer = SimpleImputer(strategy="mean")

cat_features = ["Make","Colour"]
door_features = ["Doors"]
numerical_features = ["Odometer (KM)"]

imputer = ColumnTransformer([("cat_imputer",cat_imputer,cat_features),("door_imputer",door_imputer,door_features),("numerical_imputer",numerical_imputer,numerical_features)],remainder="passthrough")
X_miss_transformed = imputer.fit_transform(X_misss)
X_miss_transformed
```
